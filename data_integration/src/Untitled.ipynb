```python
df['subject_armed'] = df['subject_armed'].astype(object)

df['subject_armed']
```

```python
df['firearm_reloaded'] = df['firearm_reloaded'].astype(bool)
df['sight_used'] = df['sight_used'].astype(bool)
df['officer_appointed_date'] = pd.to_datetime(df['officer_appointed_date'])
df['officer_birth_year'] = df['officer_birth_year'].astype(int)
df['status_datetime'] = pd.to_datetime(df['status_datetime'])
```

```python
df['trr_datetime'] = pd.to_datetime(df['trr_datetime'], utc=True)
df['beat'] = df['beat'].astype(int)
# df[df['officer_appointed_date'] == 'REDACTED' ] = df[df['officer_appointed_date'] == 'REDACTED' ].fillna(0)
# df[df['officer_appointed_date'] != 'REDACTED' ] = pd.to_datetime(df[df['officer_appointed_date'] != 'REDACTED' ], utc=True)
#df[df['officer_appointed_date'] != 'REDACTED' ]['officer_appointed_date'] =pd.to_datetime(df[df['officer_appointed_date'] != 'REDACTED' ]['officer_appointed_date'])
df['officer_birth_year'] = df['officer_birth_year'].fillna(0.0).astype(int)
df['officer_age'] = df['officer_age'].astype(int)
df['officer_on_duty'] = df['officer_on_duty'].astype(bool)
df['officer_injured'] = df['officer_injured'].astype(bool)
df['officer_in_uniform'] = df['officer_in_uniform'].astype(bool)
df['subject_birth_year'] = df['subject_birth_year'].astype(int)
df['subject_age'] = df['subject_age'].astype(int)
df['subject_armed'] = df['subject_armed'].astype(bool)
df['subject_injured'] = df['subject_injured'].astype(bool)
df['subject_alleged_injury'] = df['subject_alleged_injury'].astype(bool)
df['notify_oemc'] = df['notify_oemc'].astype(bool)
df['notify_district_sergeant'] = df['notify_district_sergeant'].astype(bool)
df['notify_op_command'] = df['notify_op_command'].astype(bool)
df['notify_det_division'] = df['notify_det_division'].astype(bool)
df['trr_created'] = pd.to_datetime(df['trr_created'], utc=True)

```

```python
df[''] = df[''].astype()
```

```python
appoint_date = df['officer_appointed_date'].to_numpy()
for i in range(len(appoint_date)):
    day = df['officer_appointed_date'][i]
    if day == 'REDACTED':
        appoint_date[i] = np.nan  
    elif len(day) == 8:
        if int(day[0:2])+2000 > 2021:
            appoint_date[i] = (19+day)
```
